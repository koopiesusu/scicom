```python
# ubah ke array
import numpy as np
import matplotlib.pyplot as plt
x = np.array(x)
y = np.array(y)

A = np.vstack([x, np.ones(len(x))]).T
y = y[:,np.newaxis]
```

```python
# step nya langsung di tumpuk
inv = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), y)

inv1 = np.dot(A.T, A)
inv2 = np.linalg.inv(inv1)
inv3 = np.dot(inv2, A.T)
inv4 = np.dot(inv3, y)

pseudo = np.linalg.pinv(A) 
pinv = np.dot(pseudo, y)
pinv

print('hasil inv', inv)
print('hasil inv', inv4) 
```
