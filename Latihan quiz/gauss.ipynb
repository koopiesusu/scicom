```python
import numpy as np

def diagonaly_dominan(matrix):
    matrix = np.array(matrix)
    absolutemat = np.abs(matrix)
    diag = np.diag(matrix)
    np.fill_diagonal(absolutemat, 0)
    offdiag = np.sum(absolutemat, axis=1)
    return np.all(diag >= offdiag)

def gauss(x,y, epsillon = 0.01):
    if not diagonaly_dominan:
        print('matrix is not diagonaly dominan')
        return
    
    x= np.array(x)
    y = np.array(y)
    diag = np.diag(np.array(x))
    np.fill_diagonal(x, 0)
    
    res = np.zeros(diag.shape)
    
    maxiter = 100
    for iter in range(maxiter):
        newres = np.array(res)
        for i, row in enumerate(x):
            newres[i] = (y[i] - np.dot(row, newres)) / diag[i]
            print(f'iteration {iter}: {newres}')
        
        diff = np.sqrt(np.dot(newres - res, newres - res))
        if diff <= epsillon:
            print('convergen solution: ', newres)
            return
        
        res = newres
        
    print('solution after max iteration: ', res)


# for i in range(len(ys)):
#     print(f'systen {i+1}')
#     gauss(xs[i], ys[i])
#     print('')

print('original system')
gauss(x, y)
# ini salah gaes
        
```
